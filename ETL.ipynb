```python
# importing dependencies

import pandas as pd
```

# Cleaning of car_data.csv


```python
# Read Csv file

car_data = pd.read_csv("Resources/car_data.csv")
car_data.head()
```

```python
# Keep only columns we need

car_df = car_data[["Make", "Model", "Year"]]
car_df.head(15)
```

```python
# Our goal is to find cars 2015, 2016, 2017 years of manufacture

updated_car_df = car_df[(car_df.Year == 2015) | (car_df.Year == 2016) | (car_df.Year == 2017)]
updated_car_df
```

```python
# Drop duplicates to show all unique Models of each car manufactured in 2015, 2016 and 2017 years

updated_car_database = updated_car_df[["Make"]]
final_car_database = updated_car_database.drop_duplicates(['Make'], keep = 'last').sort_values(["Make"], ascending= True)
final_car_database

```

```python
missing_data = pd.DataFrame({"Make": ["Jaguar", "Jeep", "MINI", "Smart"]})
final_car_updated = final_car_database.append(missing_data)
```

```python
make_df = final_car_updated.sort_values(["Make"], ascending = True)

make = make_df[make_df.Make != "Lotus"]
make = make[make.Make != "Scion"]

make

```

# Cleaning of used_cars.csv

```python
# Reading used_cars.csv file

used_car_data = pd.read_csv("Resources/used_cars.csv", error_bad_lines=False, warn_bad_lines = False)
used_car_data.head(20)
```

```python
# Keep only columns we need

used_car_df = used_car_data[["Make", "Model", "Year", "Price"]]
used_car_df.head()

```

# Data Cleaning


```python
# remove all rows with 99999 price 

updated_used_df = used_car_df[used_car_df.Price != 99999]
```

```python
# Rename incorrect Makes

updated_used_df["Make"] = updated_used_df["Make"].replace({"Alfa":"Alfa Romeo", "Aston": "Aston Martin", "INFINITI": "Infiniti",
                                                          "Land": "Land Rover", "Ram": "Dodge", "smart": "Smart"})

updated_used_df
```

```python
# remove all rows with "Freightliner" and "Lotus" in used car dataframe

final_changes_used = updated_used_df[updated_used_df.Make != "Freightliner"]
final_changes_used = final_changes_used[final_changes_used.Make != "Lotus"]
```

# AVG 2015


```python
# Retrieving only 2015 data

updated_used_car_dataframe_2015 = final_changes_used[(final_changes_used.Year == 2015)]
updated_used_car_dataframe_2015

```

```python
# Counting average price for each Make manufactured in 2015

avg_price_2015 = updated_used_car_dataframe_2015.groupby("Make").mean()
avg_price_2015

average_price_2015 = avg_price_2015.reset_index()
average_price_2015
```

# AVG 2016

```python
# Retrieving only 2016 data

updated_used_car_dataframe_2016 = final_changes_used[(final_changes_used.Year == 2016)]
updated_used_car_dataframe_2016
```

```python
# Counting average price for each Make manufactured in 2016

avg_price_2016 = updated_used_car_dataframe_2016.groupby("Make").mean()
avg_price_2016

average_price_2016 = avg_price_2016.reset_index()
average_price_2016
```

# AVG 2017

```python
# Retrieving only 2017 data

updated_used_car_dataframe_2017 = final_changes_used[(final_changes_used.Year == 2017)]
updated_used_car_dataframe_2017
```

```python
# Counting average price for each Make manufactured in 2017

avg_price_2017 = updated_used_car_dataframe_2017.groupby("Make").mean()
avg_price_2017

average_price_2017 = avg_price_2017.reset_index()
average_price_2017
```

# Fixing and Representing all tables we will use for SQL

```python
# Renaming columns of our dataframes

make = make.rename(columns = {"Make": "make"})
average_price_2015 = average_price_2015.rename(columns = {"Make": "make", "Year": "year", "Price": "price"})
average_price_2016 = average_price_2016.rename(columns = {"Make": "make", "Year": "year", "Price": "price"})
average_price_2017 = average_price_2017.rename(columns = {"Make": "make", "Year": "year", "Price": "price"})
```

```python
# Adding unique id column to our dataframes

make['id'] = range(1, 1+len(make))
updated_make = make[["id", "make"]]

average_price_2015['id'] = range(1, 1+len(average_price_2015))
updated_average_price_2015 = average_price_2015[["id", "make", "year", "price"]]

average_price_2016['id'] = range(1, 1+len(average_price_2016))
updated_average_price_2016 = average_price_2016[["id", "make", "year", "price"]]

average_price_2017['id'] = range(1, 1+len(average_price_2017))
updated_average_price_2017 = average_price_2017[["id", "make", "year", "price"]]

```

```python
# Had to make some more cleaning
# Removing "Genesis" Make

final_make = updated_make[updated_make.make != "Genesis"]
final_make
```

```python
# Removing "Scion" Make

final_average_price_2015 = updated_average_price_2015[updated_average_price_2015.make != "Scion"]
final_average_price_2015
```

```python
# Removing "Scion" Make

final_average_price_2016 = updated_average_price_2016[updated_average_price_2016.make != "Scion"]
final_average_price_2016
```

```python
# Removing "Genesis" Make

final_average_price_2017 = updated_average_price_2017[updated_average_price_2017.make != "Genesis"]
final_average_price_2017
```

# Connecting to SQL

```python
# Import dependencies

from sqlalchemy import create_engine
```

```python
# creating connection

connection_string = "postgres:postgres@localhost:5432/avg_car_prices"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables

engine.table_names()
```

# Load DataFrames into database


```python
final_make.to_sql(name = 'make', con = engine, if_exists = 'append', index = False)
```

```python
final_average_price_2015.to_sql(name = 'avg_price_2015', con = engine, if_exists = 'append', index = False)
```

```python
final_average_price_2016.to_sql(name = 'avg_price_2016', con = engine, if_exists = 'append', index = False)
```

```python
final_average_price_2017.to_sql(name = 'avg_price_2017', con = engine, if_exists = 'append', index = False)
```

# Visualization

```python
# Import dependencies

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Creating bar to represesnt all Makes manufactured in 2015 and their average price 

plt.figure(figsize = (20, 12))

y = final_average_price_2015["price"]
x = final_average_price_2015["make"]

plt.bar(x, y, color = "blue", alpha = 0.5, align = "center")

tick_locations = [value for value in x]
plt.xticks(tick_locations, rotation = 90, size = 15)

plt.xlim(-0.75, len(x) - 0.25)
plt.ylim(0, max(y) + 10000)

plt.title("AVG Price 2015", size = 25)
plt.xlabel("Make of car", size = 20)
plt.ylabel("Average Price", size = 20)


# plt.savefig("Images/AVG_price_2015.png")

plt.show()

```

```python
# Creating bar to represesnt all Makes manufactured in 2016 and their average price 

plt.figure(figsize = (20, 12))

y = final_average_price_2016["price"]
x = final_average_price_2016["make"]

plt.bar(x, y, color = "red", alpha = 0.5, align = "center")

tick_locations = [value for value in x]
plt.xticks(tick_locations, rotation = 90, size = 15)

plt.xlim(-0.75, len(x) - 0.25)
plt.ylim(0, max(y) + 10000)

plt.title("AVG Price 2016", size = 25)
plt.xlabel("Make of car", size = 20)
plt.ylabel("Average Price", size = 20)

# plt.savefig("Images/AVG_price_2016.png")

plt.show()
```

```python
# Creating bar to represesnt all Makes manufactured in 2017 and their average price 

plt.figure(figsize = (20, 12))

y = final_average_price_2017["price"]
x = final_average_price_2017["make"]

plt.bar(x, y, color = "grey", alpha = 0.5, align = "center")

tick_locations = [value for value in x]
plt.xticks(tick_locations, rotation = 90, size = 15)

plt.xlim(-0.75, len(x) - 0.25)
plt.ylim(0, max(y) + 10000)

plt.title("AVG Price 2017", size = 25)
plt.xlabel("Make of car", size = 20)
plt.ylabel("Average Price", size = 20)

# plt.savefig("Images/AVG_price_2017.png")

plt.show()
```

```python
# Combined bar, that includes all 3 bars above

make_types = final_make["make"]
n_groups = 40
fig, ax = plt.subplots()
fig.set_size_inches((20, 12))
index = np.arange(n_groups)
bar_width = 0.25
opacity = 0.8

plt.bar(x = index, height = final_average_price_2015["price"], width = bar_width, color = "blue", label = "2015")

plt.bar(x = index + bar_width, height = final_average_price_2016["price"], width = bar_width, color = "red", label = "2016")

plt.bar(x = index + 2*bar_width, height = final_average_price_2017["price"], width = bar_width, color = "grey", label = "2017")

plt.title("Comparison of average price by make", size = 20)
plt.ylabel("Average Price", size = 15)
plt.xlabel("Make of car", size = 15)
plt.xticks(index + bar_width, make_types, rotation = 90, size = 15)
plt.legend(prop = {"size":15})

plt.tight_layout()

# plt.savefig("Images/Comparison_of_avg_price.png")

plt.show()
```

```python
# ANOVA statistical test to check is it any significant difference between all average prices 

from scipy.stats import stats

stats.f_oneway(final_average_price_2015["price"], final_average_price_2016["price"], final_average_price_2017["price"])
```
